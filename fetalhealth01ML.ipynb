```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(sampling_strategy='not majority')
x_over,y_over=ros.fit_resample(x,y)
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x_scaled=std.fit_transform(x_over)
x_scaled=pd.DataFrame(data=x_scaled,columns=x_over.columns)
x_scaled.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
x_pca=pca.fit_transform(x_scaled)
x_pca=pd.DataFrame(x_pca)
x_pca.head()
```

```python
import plotly.express as px
fig = px.scatter_3d(x_pca, x=0, y=1, z=2,
              color=y_over, size_max=10)
fig.show()
```

```python
from sklearn .decomposition import PCA
pca_final=PCA(n_components=0.95)
pca_final=pca_final.fit_transform(x_scaled)
pca_final=pd.DataFrame(pca_final)
pca_final.head()
```

```python
from sklearn .decomposition import PCA
pca_final=PCA(n_components=0.85)
pca_final=pca_final.fit_transform(x_scaled)
pca_final=pd.DataFrame(pca_final)
pca_final.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(pca_final,y_over,test_size=0.25,random_state=421)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
knn_selection=[]
for k in range(1,20):
  knn=KNeighborsClassifier(n_neighbors=k)
  score=cross_val_score(knn,X_train,Y_train,cv=5)
  knn_selection.append(score.mean())
print(knn_selection)

xticks=[i for i in range(1,20)]
plt.plot([k for k in range(1,20)],knn_selection)
plt.xticks(ticks=xticks)
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,Y_train)
from sklearn.metrics import confusion_matrix
confusion_knn=confusion_matrix(Y_test,knn.predict(X_test))
snb.heatmap(confusion_knn,annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,knn.predict(X_test)))
```

```python
from sklearn import metrics
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_knn, display_labels = [False, True])
```

```python
cm_display.plot()
plt.show()
```

```python
print("Precision Score : ",metrics.precision_score(Y_test,knn.predict(X_test),average='micro'))
print("Recall Score : ",metrics.recall_score(Y_test,knn.predict(X_test),average='micro'))
print("F1_score : ",metrics.f1_score(Y_test,knn.predict(X_test),average='micro'))
#F1_score = metrics.f1_score(Y_test,knn.predict(X_test))
```

```python
param_grid={'C':[0.001,0.01,0.1,1,10,100],'gamma':[0.001,0.01,0.1,1,10,100]}
svm=SVC()
rcv=RandomizedSearchCV(svm,param_grid,cv=5)
rcv.fit(X_train,Y_train)
rcv_pred=rcv.predict(X_test)
CM_svm=confusion_matrix(Y_test,rcv_pred)
print(CM_svm)
snb.heatmap(CM_svm,annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,rcv_pred))
```

```python
#precison, sensitivity or recall and F1-Score for SVM
print("Precision Score : ",metrics.precision_score(Y_test,rcv_pred,average='micro'))
print("Recall Score : ",metrics.recall_score(Y_test,rcv_pred,average='micro'))
print("F1_score : ",metrics.f1_score(Y_test,rcv_pred,average='micro'))
```
